```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns; sns.set()
import cv2
import os
from matplotlib import pyplot as plt
from matplotlib import animation
from JSAnimation import IPython_display
```

```python
def get_Images(path):
    i = 0
    img_struct = []
    for (path, dirs, files) in os.walk(path):
        for fname in files:
            if fname.endswith('jpg'):
                src = os.path.join(path, fname)
                #print(src)
                img = cv2.imread(src, 0)
                d1, d2 = img.shape
                img = img.flatten()
                img_struct.append(img)
    return d1,d2,img_struct
```

# K Nearest Neighbours

```python
sel_TestWeights = weights_Test[1,:]
err = np.sqrt(np.sum(np.square(weights_Train - sel_TestWeights[np.newaxis, :]), axis = 1))
len(err)
```

```python
err = np.argsort(err)
```

```python
err[err <= (3868/2) - 1] = -1
err[err >= (3868/2)] = 1
```

```python
np.any(err > 1)
```

```python
fin_Class = np.sum(err[0:10])
fin_Class
```

Note that __.append()__ will return an empty object. So something like __A = A.append(i)__ will just return an empty object. We just need to do __A.append()__. 

```python
def k_NN(weights_Test, weights_Train, k):
    miss_ClassM = []
    miss_ClassF = []
    for NN in k:
        k_NNMale = []
        k_NNFemale = []
        for i in range(len(weights_Test[:,1])):
            sel_TestWeights = weights_Test[i,:]
            err = np.sqrt(np.sum(np.square(weights_Train - sel_TestWeights[np.newaxis, :]), axis = 1))

            err = np.argsort(err)

            err[err <= (3868/2) - 1] = -1
            err[err >= (3868/2)] = 1

            fin_Class = np.sum(err[0:NN])

            if i <= (2000/2) - 1:
                k_NNMale.append(fin_Class)
            else:
                k_NNFemale.append(fin_Class)

        k_NNMale = [0 if p <= 0 else 1 for p in k_NNMale]
        k_NNFemale = [1 if p < 0 else 0 for p in k_NNFemale]

        err_Male = sum(k_NNMale)
        err_Female = sum(k_NNFemale)
        
        miss_ClassM.append(err_Male)
        miss_ClassF.append(err_Female)
    
    return miss_ClassM, miss_ClassF
```

```python
k = [1, 5, 10, 50, 100]
miss_ClassM, miss_ClassF = k_NN(weights_Test, weights_Train, k)
```

```python
ind = np.array([1,2,3,4,5])  # the x locations for the groups
width = 0.27 # the width of the bars

fig = plt.figure(facecolor = 'white')
ax = fig.add_subplot(111)

yvals = [(i/1000)*100 for i in miss_ClassM]
rects1 = ax.bar(ind, yvals, width)
zvals = [(i/1000)*100 for i in miss_ClassF]
rects2 = ax.bar(ind+width, zvals, width)

ax.set_ylabel('Misclassified Values')
ax.set_xlabel('K - Nearest Neighbours')
ax.set_xticks(ind+ width/2)
ax.set_xticklabels( ('1', '5', '10', '50', '100') )
ax.legend( (rects1[0], rects2[0]), ('Male', 'Female') )
```
