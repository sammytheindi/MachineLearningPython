The first thing that we will do is centralize the data. We will do this by taking the mean of each pixel, and then subtracting that mean from each face in the training data. The documentation for mean is given [here](https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.mean.html). Note that we converted the array to a floating point value so that the mean calculations are more accurate. 

Now note that taking a transpose of a 1D array will just give a 1D array, and we will have to do broadcasting to ensure that we can subtract this mean from each image. So we look at our dimensions. The dimensions of the faces matrix is __[10304, 360]__. The dimensions of the face_means matrix is __[10304]__. As such, since according to the rules of broadcasting all of the trailing axes need to either match or be 1, we will add a new axis to the right so that face_means becomes __[10304, 1]__. Now when we do our subtration operation, the face_means matrix will be stretched over all the different faces in the faces matrix.

```python
face_means = np.mean(img_struct, axis = 1)
face_means = np.transpose(face_means)

face_means_s = face_means.reshape(d1,d2)
plt.imshow(face_means_s, cmap = plt.get_cmap('gray'))
plt.xticks([]), plt.yticks([])
plt.show
```

```python
face_means.shape
```

```python
face_means = face_means[:, np.newaxis]
```

Now note the shape of the face_means matrix

```python
face_means.shape
```

So if we now subtract the mean, we will get the following

```python
c_faces = img_struct - face_means
c_faces.shape
```

```python
corr_faces = np.matmul(np.transpose(c_faces), c_faces) 
corr_faces.shape
```

```python
val, vec = np.linalg.eig(corr_faces)
val = np.absolute(val)
```

```python
xi = [i for i in range(0, len(val))]
plt.bar(xi, val)
```

```python
print(vec.shape)
print(c_faces.shape)
```

```python
eigFaces = np.matmul(c_faces, vec)
```

```python
norm_eigFaces = eigFaces / np.sqrt(val[np.newaxis, :])
```

```python
norm_eigFaces.shape
```

```python
fig, axes = plt.subplots(1,4, figsize=(18, 16))
for k in range(4):
    eigFace = norm_eigFaces[:,k].reshape(d1,d2)
    
    axes[k].set_xticks([]) 
    axes[k].set_yticks([])
    axes[k].set_title('EigenFace ' + str(k+1))
    axes[k].imshow(eigFace, cmap = plt.get_cmap('jet'))
```

```python
weights = np.dot(np.transpose(c_faces), norm_eigFaces)
weights.shape
```

```python
rec_img = np.zeros(d1*d2)
rec_img.shape
```

```python
rec_img = weights[0,:][np.newaxis, :]*norm_eigFaces
print(rec_img)
```

```python
rec_img.shape
```

```python
rec_img.sum(axis=1).shape
```

```python
fig, axes = plt.subplots(3,3, figsize=(18, 16))
axes = axes.flatten()

newimg = rec_img[:,0]

axes[0].set_xticks([]) 
axes[0].set_yticks([])
axes[0].set_title(str(1) + ' Eigenfaces')
axes[0].imshow(newimg.reshape(d1,d2), cmap = plt.get_cmap('gray'))

i = 1;
for k in range(39,359, 40):
    newimg = rec_img[:,0:k].sum(axis = 1)
    
    axes[i].set_xticks([]) 
    axes[i].set_yticks([])
    axes[i].set_title(str(k+1) + ' Eigenfaces')
    axes[i].imshow(newimg.reshape(d1,d2), cmap = plt.get_cmap('gray'))
    i = i+1
    
    
```
